Ticks, Labels and Legends

```python
fig= plt.figure()
ax= fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ticks= ax.set_xticks([0,250,500,750,1000])
labels=ax.set_xticklabels(['1','2', '3', '4','5'], rotation= 30, fontsize= 'small')
ax.set_title('usdkjlh,')
ax.set_xlabel('stages')

```

I like it

```python
# doing everything all at once
fig= plt.figure()
ax= fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k--')

props={
    'title': 'This I like'
    ,'xticks': [0,250,500,750,1000]
    ,'xticklabels': ['a', 'b','c','d','e']
    ,'xlabel': 'stages'
    ,'yticks': [-20,-10, 0, 10, 20 ]
    ,'yticklabels': ['a', 'b','c','d','e']
    ,'ylabel': 'y_stage'
}

ax.set(**props)
```

```python
fig= plt.figure()
ax= fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), linestyle= '--', color= 'g', label= 'one')
ax.plot(np.random.randn(1000).cumsum(), linestyle= ':', color= 'b', label= 'two')
ax.plot(np.random.randn(1000).cumsum(), linestyle= '-.', color= 'c', label= 'three')
ax.legend(loc= 'best')

props={
    'title': 'This I like'
    ,'xticks': [0,250,500,750,1000]
    ,'xticklabels': ['a', 'b','c','d','e']
    ,'xlabel': 'stages'
    ,'yticks': [-20,-10, 0, 10, 20 ]
    ,'yticklabels': ['a', 'b','c','d','e']
    ,'ylabel': 'y_stage'
}

ax.set(**props)
```
